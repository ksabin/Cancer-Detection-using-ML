#### 7.1.1 Removing Outliers in Continuous columns

##### 7.1.1.1 IQR method for Outliers Removal

```python
df_cleaned = data.copy()
def replace_outliers(df, column):
    #Step-1: Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    #Step-2: Calculate IQR
    IQR = Q3 - Q1
    #Step-3: Calculate lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    #Step-4: Replace outliers below lower bound with Q1
    data[column]= data[column].apply(lambda x: Q1 if x < lower_bound else x)
    #Step-5: Replace outliers above upper bound with Q3
    data[column]= data[column].apply(lambda x: Q3 if x > upper_bound else x)
    return data

#For removing outliers of Multiple columns
for col in continuous_col:
    df_cleaned = replace_outliers(df_cleaned, col)
```

# 8. Feature Selection

### 8.1 ANOVA F-TEST


Insights: 
* ANOVA = Analysis of Variance.
* Measures how much the mean of each feature varies across classes (benign vs malignant).
* If a feature's values are very different for the two classes, it's likely a good predictor.
* f_classif is used for numerical features and categorical target.

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Split features and target
a = data.drop(['id', 'diagnosis'], axis=1)
b = data['diagnosis']

# Apply SelectKBest with ANOVA F-test
anova_selector = SelectKBest(score_func=f_classif, k='all')
x_new = anova_selector.fit_transform(a, b)

# Get F-scores and feature names
f_scores = anova_selector.scores_
f_scores_df = pd.DataFrame({
    'Feature': a.columns,
    'F-score': f_scores
})
```

```python
f_scores_df.sort_values(by='F-score', ascending = False).reset_index(drop=True)
```

Insights:
* F = 0 to 1 ---> likely not useful and consider dropping
* F = 1 to 10 ---> Use with caution, weak to moderate relationship
* F = 10 to 100 ---> Likely useful and consider selecting
* F > 100 ---> Very important and has very strong discrimination

#### 11.1.1 KNNClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

About KNN:
* When you want to classify or predict something, KNN looks at the closest data points (neighbors) to it.
* It compares the input data to the existing data and decides based on whatâ€™s most common among its nearest neighbors.

What KNN Means
* K = The number of neighbors you want to look at
* Nearest = The closest or most similar items
* Neighbors = The data points around your new item

How It Works
* You have a dataset with known items (like fruits with their names)
* When a new item comes in, KNN:
* Measures how similar it is to all existing items
* Finds the K closest matches
* Takes a "vote" from these neighbors
* Makes its decision based on the majority

# 13. Model Evaluation

Insights: 
* For Classification problem, we use Confusion Matrix, Precision, Recall, F1 score, accuracy for model evaluation

### 13.1 Confusion Matrix For KNNClassifier

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#Calculate confusion matrix for logistic regression
cm_knn = confusion_matrix(y_test, prediction_knn)
print("Confusion Matrix for Logistic Regression:\n", cm_knn)

#Visualize confusion matrix
disp_knn = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=['class 0', 'class 1'])
disp_knn.plot(cmap=plt.cm.Blues)
plt.show()

```

Insight:
* class=0 --> Benign

* class=1 --> Malignant

* If predicted= actual --> True, else --> False

* if predicted value is: 0= Negative and 1= Positive

* True Negative, TN=70 ; False Positive, FP=1

* False Negative, FN=2 ; True Positive, TP=41

#### 13.1.1 Calculation of Precision, Accuracy, F1-score and Recall Score For KNN classifier

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
precision= precision_score(y_test, prediction_knn)
recall_score= recall_score(y_test, prediction_knn)
accuracy_score= accuracy_score(y_test, prediction_knn)
f1_score= f1_score(y_test, prediction_knn)
print("KNNClassifier Performance Metrics:")
print(f"Precision: {precision:.2f}")   
print(f"Recall: {recall_score:.2f}")
print(f"Accuracy: {accuracy_score:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```
